Load the model with sentence-transformers. Try an example.

```python
from sentence_transformers import SentenceTransformer, util
sentence1 = ["This is an example sentence"]
sentence2 = ["Each sentence is converted"]

model = SentenceTransformer("neuml/pubmedbert-base-embeddings")
embeddings1 = model.encode(sentence1)
embeddings2 = model.encode(sentence2)
print("Embeddings 1")
print(embeddings1)
print("Embeddings 2")
print(embeddings2)
```

Results of the example:

```python
cosine_scores = util.cos_sim(embeddings1, embeddings2)
print(cosine_scores[0][0])
```

Second, load data as dataframe from a csv file.

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/BİL495 496/Data/data.csv')
df.head()
```

Organize data accordingly.

```python
q1_list = df['S1'].values.tolist()
print(q1_list)

q2_list = df['S2'].values.tolist()
print(q2_list)
```

Give sentences to model. Find sentence similarities.

```python
model = SentenceTransformer("neuml/pubmedbert-base-embeddings")

embeddings1 = model.encode(q1_list)
embeddings2 = model.encode(q2_list)

print("Embeddings 1")
print(embeddings1)
print("Embeddings 2")
print(embeddings2)
```

```python
cosine_scores = util.cos_sim(embeddings1, embeddings2)
results = []

for i in range(len(q1_list)):
  print(cosine_scores[i][i])
  result = []
  result.append(q1_list[i])
  result.append(q2_list[i])
  result.append(cosine_scores[i][i])
  results.append(result)

print(results)
```

Write results in a different csv file (here, "results.csv").

```python
import csv
with open('/content/drive/MyDrive/BİL495 496/Data/results.csv', 'w', newline='') as file:
    # Step 4: Using csv.writer to write the list to the CSV file
    writer = csv.writer(file)
    writer.writerows(results)
```
